## 正则化线性回归
### 查看数据

```python
data = sio.loadmat('./data/ex5data1.mat')
X = data['X']
y = data['y']
Xval = data['Xval']
yval = data['yval']
Xtest = data['Xtest']
ytest = data['ytest']

# 插入 1 行
X =     np.insert(X    ,0,1,axis=1)
Xval =  np.insert(Xval ,0,1,axis=1)
Xtest = np.insert(Xtest,0,1,axis=1)
X.shape, y.shape, Xval.shape,yval.shape,Xtest.shape,ytest.shape
```

### 正则化代价函数（线性回归）

$J(\theta)=\frac{1}{2m}\sum\limits_{i=1}^m [(h_{\theta}(x^{(i)})-y^{(i)})^2+\lambda\sum\limits_{i=1}^{n}\theta_j^2 ]\\ =\frac{1}{2m}[(X\theta-y)^T(X\theta-y)+\lambda\sum\limits_{j=1}^{n}\theta_j^2]$

```python
def computeRegCost(theta, X, y, mylambda=0.):
    theta = np.matrix(theta).T
    X = np.matrix(X)
    y = np.matrix(y)
    m = len(X)

    cost = X*theta - y
    
    left = (1./(2*m )) * np.dot(cost.T,cost)
    left = np.array(left).ravel()
    
    reg = (mylambda/(2*m)) * float(theta[1:].T *(theta[1:]))
    
    return left[0] + reg
```

```python
theta = np.array([[1.],[1.]]).T
computeRegCost(theta,X,y,mylambda=1.)
```

### 正则化梯度（线性回归）

$Repeat\ until\ convergence\{\\ \theta_0=\theta_0-\alpha\frac{1}{m}\sum\limits_{i=1}^m [(h_{\theta}(x^{(i)})-y^{(i)})x_0^{(i)}] \\ \theta_j=\theta_j-\alpha\frac{1}{m}\sum\limits_{i=1}^m [(h_{\theta}(x^{(i)})-y^{(i)})x^{(i)} + \frac{\lambda}{m}\theta_j]\\for \ j=1,2,..,n\\\}$

```python
# 计算迭代一次的梯度
def computeRegGradient(theta, X, y, mylambda=0.):
    theta = np.matrix(theta).T
    X = np.matrix(X)
    y = np.matrix(y)
    m = len(X)
    n = theta.shape[1]  
    
    temp = (1./m) * (X.T * (X*theta - y)) # 维度 2*1
    reg  = (mylambda/m) * theta
    # theta0不参与正则化
    reg[0] = 0 

    return np.array(temp+ reg).ravel()
```

```python
computeRegGradient(theta, X, y, mylambda=1.)
```

### 拟合模型

最小化参数

```python
def fitting_theta(theta,X, y,mylambda):
    result = opt.fmin_cg(f=computeRegCost, x0=theta, fprime=computeRegGradient, args=(X, y,mylambda), disp=False,full_output=True)
    return result[0],result[1]
```

```python
def fitting_theta1(theta,X, y,mylambda):
    result = opt.fmin_cg(computeRegCost,x0=theta,\
                                       fprime=computeRegGradient,\
                                       args=(X,y,mylambda),\
                                       full_output=True,\
                                       epsilon=1.49e-12,\
                                       maxiter=1000)

    return result[0].reshape(1,X.shape[1]),result[1]
```

```python
a,b= fitting_theta1(theta,X, y,1.)
theta.shape
```

```python
mylambda=1.
fit_theta,mincost= fitting_theta(theta,X, y,mylambda)
mincost
```

## 偏差与方差

### 学习曲线

```python
# 分别计算训练集和交叉验证集的误差
training_error, cv_error = [], []

m = X.shape[0]
for i in range(1, m+1):
    init_theta = np.ones((1,X.shape[1]))
    fit_theta,mincost_tr = fitting_theta(init_theta, X[:i],  y[:i],1)
    
    mincost_cv = computeRegCost(fit_theta, Xval, yval)
    training_error.append(mincost_tr)
    cv_error.append(mincost_cv)
```

```python
# 绘制学习曲线
plt.figure(figsize=(8,5))
plt.plot(np.arange(1, m+1), training_error,label='Train')
plt.plot(np.arange(1, m+1), cv_error,label='Cross Validation')
plt.legend()
plt.title('Learning curve for linear regression')
plt.xlabel('Number of training examples')
plt.ylabel('Error')
plt.grid(True)
```

### 使用多项式回归

### 创建多项式特征并标准化

```python
from sklearn import preprocessing

def genPolyFeature(X,dim):
    X_new = X.copy()
    for i in range(2,dim+1):
        p = i
        X_new = np.insert(X_new,X_new.shape[1],np.power(X_new[:,1],p),axis=1)
    
    return X_new
```

### 标准化时，顺便返回均值与方差

```python
# 标准化
def featureNormalize(myX):
    df = pd.DataFrame(myX[:,1:])
    df0 = df.apply(lambda column: (column - column.mean()) / column.std())
    myX_norm = np.matrix(df0)
    myX_norm = np.insert(myX_norm, 0 ,1,axis=1)
    # 求出均值和方差，保留以便后续标准化使用
    df1 =df.apply(lambda column: column.mean())
    means = np.array(df1).reshape(len(df1),1)
    means = means.T
    
    df2 =df.apply(lambda column: column.std())
    stds = np.array(df2).reshape(len(df2),1)
    stds = stds.T
    return myX_norm,means,stds
```

```python
dim = 8
X_new= genPolyFeature(X,dim)
Xval_new = genPolyFeature(Xval,dim)
Xtest_new = genPolyFeature(Xtest,dim)

X_norm,X_means,X_stds = featureNormalize(X_new)

Xval_norm,Xval_means,Xval_stds = featureNormalize(Xval_new)

Xtest_norm,Xtest_means,Xtest_stds = featureNormalize(Xtest_new)

X_norm.shape,Xval_norm.shape,Xtest_norm.shape
```

### 拟合多项式模型

```python
def fitting_model_and_plot(myX, myy,means,stds,mylamda=0.):
    init_theta = np.ones((1,myX.shape[1]))
    fit_polytheta,polymincost = fitting_theta1(init_theta, myX,  myy,mylamda)
    fit_polytheta = fit_polytheta.reshape(1,myX.shape[1])
    # 开始绘制图像
    x_v = np.linspace(-55,55,50)
    X_v_new = x_v.reshape(50,1)
    X_v_new = np.insert(X_v_new, 0, 1,axis=1)
    X_v_new = genPolyFeature(X_v_new,dim)
    
    # 标准化
    # 注意，这里是用训练集的均值和方差来进行标准化的。
    X_v_norm = (X_v_new[:,1:]-means) / stds
    X_v_norm = np.insert(X_v_norm, 0 ,1,axis=1)
    X_v_norm = np.matrix(X_v_norm)
    
    # 预测值
    h = np.dot(X_v_norm,fit_polytheta.T)
    plotData(X,y)
    plt.plot(x_v,h,'b--')
    
    return fit_polytheta,polymincost
```

```python
fitting_model_and_plot(X_norm, y, X_means,X_stds,mylamda=0.)
```

### 绘制学习曲线

```python
def learning_curve(X_norm,Xval_norm,y,yval, mylambda=0.):
    # 分别计算训练集和交叉验证集的误差
    polyTraining_error, polyCv_error = [], []

    m = X_norm.shape[0]
    for i in range(1, m+1):
        init_theta = np.ones((1,X_norm.shape[1]))
        fit_theta,mincost_tr = fitting_theta(init_theta, X_norm[:i],  y[:i],mylambda)
    
        mincost_cv = computeRegCost(fit_theta, Xval_norm, yval)
        polyTraining_error.append(mincost_tr)
        polyCv_error.append(mincost_cv)
    
    # 绘制学习曲线
    plt.figure(figsize=(8,5))
    plt.plot(np.arange(1, m+1), polyTraining_error,label='Train')
    plt.plot(np.arange(1, m+1), polyCv_error,label='Cross Validation')
    plt.legend()
    plt.title('Learning curve for linear regression')
    plt.xlabel('Number of training examples')
    plt.ylabel('Error')
    plt.ylim([0,100])
    plt.grid(True)
```

```python
learning_curve(X_norm,Xval_norm, y,yval, mylambda=0.)
```
